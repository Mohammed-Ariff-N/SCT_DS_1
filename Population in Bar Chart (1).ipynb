```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\User\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_38144\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)


india_row = df[df['Country Name'] == 'China']
# Data from 2015 to 2024
india_population = india_row.iloc[0, 62:70]  

years = india_population.index.astype(str)
values = india_population.values.astype(float)

plt.figure(figsize=(10, 5))
bars = plt.bar(years, values/1e6,color='blue')
for bar in bars:
    population_in_millions = bar.get_height()  
    
    
    x = bar.get_x() + bar.get_width() / 2
    y = population_in_millions + 5

    label = f'{population_in_millions:.1f}' 
    plt.text(x, y, label, ha='center')
plt.title("China's Population from 2015 to 2024")
plt.xlabel("Year")
plt.ylabel("Population(in Million)")
plt.grid(False)
plt.tight_layout()
plt.show()
```
